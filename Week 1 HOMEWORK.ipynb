```
Lotus_car = car_data[(car_data['Make']=='Lotus')]

Lotus_car = Lotus_car[['Engine HP','Engine Cylinders']]


Lotus_car= Lotus_car.drop_duplicates(  subset = ['Engine HP','Engine Cylinders'],  keep = 'last').reset_index(drop = True)

X = Lotus_car.to_numpy()

xtx= X.T@X

print (xtx)

from numpy.linalg import inv, det



ix = np.linalg.inv(xtx)

ix1= ix@X.T

y = np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])

w=ix1@y

print (w)
```
